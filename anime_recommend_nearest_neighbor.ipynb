```python
import pandas as pd
import numpy as np
```

```python
anime=pd.read_csv('anime.csv')
rating=pd.read_csv('anime_rating.csv')
```

```python
anime
```

```python
anime.info()
```

```python
anime.isnull().sum()
```

```python
anime.dropna(inplace=True)
```

```python
anime=anime.drop_duplicates(subset=['name'])
```

```python
rating=rating[rating['rating']!=-1]
```

```python
merged=rating.merge(anime,on='anime_id')
merged=merged.drop(columns=['rating_x'])
merged
```

```python
number_df=merged.groupby('name').count()['members'].reset_index().rename(columns={'members':'number'})
number_df=number_df[number_df['number']>10]
number_df
```

```python
final=merged.merge(number_df,on='name')
final
```

```python
pt=final.pivot_table(index='name',columns='user_id',values='rating_y')
pt.fillna(0,inplace=True)
pt.tail()
```

```python
pt.tail(50)
```

```python
from scipy.sparse import csr_matrix

m=csr_matrix(pt.values)
m
```

```python
from sklearn.neighbors import NearestNeighbors

model=NearestNeighbors(metric='cosine')
model.fit(m)
```

```python
m.shape
```

```python
distance,indices=model.kneighbors(pt.iloc[500].values.reshape(1,-1),n_neighbors=6)
print(distance)
print(indices)
```

```python
np.where(pt.index=='One Piece')
```

```python
d1={}
def recommend_anime(anime_name):
    index=np.where(pt.index==anime_name)[0][0]
    distance,indices=model.kneighbors(pt.iloc[index].values.reshape(1,-1),n_neighbors=6)
    for i in range(1,len(distance.flatten())):
        d1[pt.index[indices.flatten()[i]]]=distance.flatten()[i]
    print(d1)
```

```python
recommend_anime('eX-Driver')
```
