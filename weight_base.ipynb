```python
merged_df=ratings.merge(books,on='ISBN')
merged_df
```

```python
num_rating_df=merged_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'Num'},inplace=True)
num_rating_df
```

```python
average_rating=merged_df.groupby('Book-Title').mean()['Book-Rating'].reset_index()
average_rating.rename(columns={'Book-Rating':'avg'},inplace=True)
```

```python
df=num_rating_df.merge(average_rating,on='Book-Title')
df
```

```python
main_df=merged_df.merge(df,on='Book-Title').drop_duplicates('Book-Title').reset_index()
main_df.head(20)
```

Using Weighted average:

```python
main_df['Num'].quantile(0.80)
```

```python
r=main_df['avg']
v=main_df['Num']
m=main_df['Num'].quantile(0.75)
c=main_df['avg'].mean()
```

```python
main_df['weights']=((r*v)+(c*m))/(v+m)
main_df
```

```python
sorted_df=main_df.sort_values('weights',ascending=False)
sorted_df.head(10)
```
