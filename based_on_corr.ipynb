```python
import pandas as pd
```

```python
book=pd.read_csv('Books.csv')
ratings=pd.read_csv('Ratings.csv')
```

```python
book_rating=book.merge(ratings,on=['ISBN'])
book_rating.dropna(inplace=True)
book_rating
```

```python
book_rating.duplicated().sum()
```

```python
mean_rating=pd.DataFrame(book_rating.groupby('Book-Title')['Book-Rating'].mean())
```

```python
mean_rating['num_of_rating']=pd.DataFrame(book_rating.groupby('Book-Title')['Book-Rating'].count())
mean_rating
```

```python
book_pivot=book_rating.pivot_table(index='User-ID',columns='Book-Title',values='Book-Rating')
book_pivot
```
